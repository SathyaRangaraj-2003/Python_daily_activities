```python
import csv
import json
import pandas as pd
```

```python
#TASK 1 : dataloading and cleaning
#loading

df_csv = pd.read_csv('old_airline_data_2023.csv')
df_json = pd.read_json('new_airline_data_2024.json',)
print(df_json)
```

```python
#identify all missing values

missing_values = df_json.isnull().sum()
print(missing_values)
```

```python
#numeric columns

df_json['Flights Operated'] = df_json['Flights Operated'].fillna(df_json['Flights Operated'].mean())

df_json.iloc[:,2:] = df_json.iloc[:,2:].fillna(df_json.iloc[:,2:].mean())

print(df_json)
```

```python
#drop simultaneously(all three numeric columns)

df_json = df_json.dropna(subset=['Flights Operated', 'Passengers Carried' , 'Target Passengers'] , how = 'all',axis=0)
print(df_json)
```

```python
#validate the month

df_json = df_json[df_json.iloc[:,0].isin(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])]
print(df_json)
```

```python
#standardization:

df_json.columns = df_json.columns.str.lower()
df_json.columns = df_json.columns.str.replace(' ','_')
df_json['month'] = df_json['month'].str.capitalize()
print(df_json)
```

```python
#Duplicate Handling:

df_json = df_json.groupby('month',as_index = False).first()
print(df_json)
```

```python
#cleaning
df_csv.columns = df_csv.columns.str.lower().str.replace(' ','_')
df_csv
```

```python
#TASK 2 : DATA MERGING
#concatenation:

new_data = pd.concat([df_csv , df_json] , ignore_index = True)
print(new_data)
```

```python
#year_column

new_data['year'] = ["2023" if idx < len(df_csv)  else "2024" for idx in range(len(new_data))]
print(new_data)
```

```python
#month consistency:

df_json = new_data[new_data.month.isin(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])]
print(df_json)
```

```python
#sorting

#chronologically by month
month_order = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
new_data['month'] = pd.Categorical(new_data['month'],categories = month_order , ordered = True)
new_data.sort_values(['year','month'],inplace=True)
new_data
```

```python
#TASK 3
#performance metrics
new_data['pct_diff_target'] = ((new_data['passengers_carried'] - new_data['target_passengers']) / new_data['target_passengers']) * 100
new_data
```

```python
#Target Non-Achievement:

target_not_achieved = new_data[new_data['pct_diff_target'] < 0 ]
target_not_achieved[['month','year','pct_diff_target']]
```

```python
#Annual Summaries:

total_passenger = new_data.groupby('year',as_index = False)['passengers_carried'].sum().astype(int)
total_passenger
```

```python
total_flights = new_data.groupby('year',as_index = False).flights_operated.sum().astype(int)
total_flights
```

```python
average = new_data.groupby('year',as_index = False)['pct_diff_target'].mean()
average
```

```python
annual_summary = pd.concat([total_passenger , total_flights.iloc[:,1] , average.iloc[:,1]]  ,ignore_index=False, axis = 1 )
annual_summary
```

```python
#Extreme Performance

best = new_data[new_data['pct_diff_target'] == new_data['pct_diff_target'].max()][['month','year']]
worst = new_data[new_data['pct_diff_target'] == new_data['pct_diff_target'].min()][['month','year']]
print(best ,"--> Best Performance")
print(worst ,"--> Worst Performance")
```

```python
#TASK 4 : Data output and zipping
#output files

out1 = new_data.to_csv('combined_airline_data.csv',index=False)

out2 = target_not_achieved.to_csv('target_not_achieved.csv' ,  index = False)
out3 = annual_summary.to_json('yearly_summary.json')
```
